```python
#### takes images: "IMG_1357.jpg","IMG_1358.jpg",.... and extract mean RGB,HSV, LAB values of reference zone and sample zone

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import cv2
import pandas as pd 
import scipy.io
```

```python
##### code to extract mean rgb,hsv,lab values from all images and store them in a csv file #######

rgb_list=[]
for i in range (1,602):
    if i >0:

        ## path to the Image dataset directory 
        filename= "C:/Users/Brinda Khanal/Desktop/All pesticide test data/combined/IMG_"+str(i)+".jpg"
        print (filename) 

        image_name= ["IMG_"+str(i)]

        ## read the image
        image = cv2.imread(filename)
        #image_for_plot= np.copy(image)
        image_for_plot= cv2.cvtColor(image,cv2.COLOR_BGR2LAB)


        # rgb to gray conversion
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

        gray_mean= np.mean(gray)
        print (gray_mean)

        #formula to choose right lower binary threshold, self-estimated formula after observance 
        # T= 0.8I-30
        lower_th= 0.8*gray_mean-30

        # threshold, gives binary image
        ret2,th2 = cv2.threshold(gray,lower_th,255,cv2.THRESH_BINARY)


        # find the contour 
        contours= cv2.findContours(th2,cv2.RETR_CCOMP,cv2.CHAIN_APPROX_NONE)
        cnts = contours[1] if cv2.__version__.startswith('3') else contours[0]
        circles= []

        for c in cnts:

            if cv2.contourArea(c) < 2000:
                continue

            (x,y),r = cv2.minEnclosingCircle(c) # get the minimum closing circle
            circles.append([int(x),int(y),int(r)-25])

        # re-arrange the circles from left to right (i.e with the horizontal)
        if circles[0][0]>circles[1][0]:
            circles= circles[0:2][::-1]



        RGB_values=[]  # stores RGB value of reference zone and sample zone

        for cr in circles[0:2]:
            mask = np.zeros(image.shape,np.uint8) # empty array of image size
            cv2.circle(mask,(cr[0],cr[1]),cr[2],(255,255,255),-1) # create a circular mask

            masked_image = cv2.bitwise_and(image_for_plot,mask) # masked image

            pixelpoints = masked_image[~np.all(masked_image == 0, axis=2)] #obtain the RGB,HSV,LAB values of pixel points inside the circle
            RGB_values.append(pixelpoints)

        RGB_values= np.array(RGB_values) 
        mean_ref = np.mean(RGB_values[0], axis=0)
        print (mean_ref)
        mean_sample=np.mean(RGB_values[1], axis=0)
        concat_data=np.concatenate((image_name,mean_ref,mean_sample), axis=0)
        #rgb_list.append(image_name)
        rgb_list.append(concat_data)
        
rgb_list=list(rgb_list)         
#df=pd.DataFrame(rgb_list, columns=["ImageName","BlueReference","GreenReference","RedReference","BlueSample","GreenSample","RedSample"] )
#df=pd.DataFrame(rgb_list, columns=["ImageName","HReference","SReference","VReference","HSample","SSample","VSample"])
df=pd.DataFrame(rgb_list, columns=["ImageName","LReference","aReference","bReference","LSample","aSample","bSample"])
Label_data= pd.read_excel("C:/Users/Brinda Khanal/Desktop/All pesticide test data/combined/Labels.xlsx")
result = df.merge(Label_data, on ='ImageName')
result.to_csv("C:/Users/Brinda Khanal/Documents/Bidur Git Repo/padColorimetry/Data/new data/mean_pixel_LAB_pesticide_test_new.csv")
result.head(30)
```
