# This code is used read the csv file and produce the baxplot of the values #

```python
def plot_data(data, data2):
    
    color_channels= data['remarks'].str.split(",").str.get(0)
    data=pd.concat([data,data2])
    data= data.dropna() 
    print (data)
     
    plot_data=data.melt(id_vars=['remarks','avg_accuracy','std'])
    plot_data['color_channel']= plot_data['remarks'].str.split(",").str.get(0)
    plot_data['Models']= plot_data['remarks'].str.split(",").str.get(1)
    plot_data['Models'].replace({"Logistic Regression": "LR", "Random Forest": "RF", "CNN": "ANN"}, inplace=True)


    plot_data= plot_data.rename(columns=color_channels)
    return plot_data
```

```python

params = {'legend.fontsize': 'xx-large',
          'figure.figsize': (20, 7),
         'axes.labelsize': 'xx-large',
         'axes.titlesize':'xx-large',
         'xtick.labelsize':'xx-large',
         'ytick.labelsize':'xx-large'}

plt.rcParams.update(params)


plt.subplots_adjust(left=0.125, bottom=-0.1, right=0.9, top=1, wspace=0.2, hspace=1)
plt.subplot(221)

# for the train data
pca = PCA(2)  # project from 6 to 2 dimensions
projected = pca.fit_transform(data1)
print(data1.shape)
print(projected.shape)
plt.scatter(data1.values[:, 4], data1.values[:, 5],
            c=food_color_train_data['Dilution Factor'], edgecolor='none', alpha=1,
            cmap=plt.cm.get_cmap('Spectral', 10))
plt.title('Food Color, Training Set')
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.colorbar();

plt.subplot(222)

# for the train data
pca = PCA(2)  # project from 6 to 2 dimensions
projected = pca.fit_transform(data2)
print(data2.shape)
print(projected.shape)
plt.scatter(data2.values[:, 4], data2.values[:, 5],
            c=food_color_test_data['Dilution Factor'], edgecolor='none', alpha=1,
            cmap=plt.cm.get_cmap('Spectral', 10))
plt.title('Food Color Test Set')
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.colorbar();

plt.subplot(223)

# for the train data
pca = PCA(2)  # project from 6 to 2 dimensions
projected = pca.fit_transform(data3)
print(data3.shape)
print(projected.shape)
plt.scatter(data3.values[:, 4], data3.values[:, 5],
            c=pesticide_train_data['Dilution Factor'], edgecolor='none', alpha=1,
            cmap=plt.cm.get_cmap('Spectral', 10))
plt.title('Pesticide Training Set')
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.colorbar();

plt.subplot(224)

# for the train data
pca = PCA(2)  # project from 6 to 2 dimensions
projected = pca.fit_transform(data4)
print(data4.shape)
print(projected.shape)
plt.scatter(data4.values[:, 4], data4.values[:, 5],
            c=pesticide_test_data['Dilution Factor'], edgecolor='none', alpha=1,
            cmap=plt.cm.get_cmap('Spectral', 10))
plt.title('Pesticide Test Set')
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.colorbar();
```
